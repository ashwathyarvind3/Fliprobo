# Question-1
Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ You need to find following details:
            A) Rank B) Name C) Artist D) Upload date E) Views

```python
# Opening the url
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/')
```

```python
#clicking the search list
search_button=driver.find_element_by_xpath("//a[@class='external text']")
search_button.click()
```

```python
#clicking the first search result 
list_search=driver.find_element_by_xpath("//ul[@class='mw-search-results']/li[1]/div/a")
list_search.click()
```

```python
# creating empty lists for scraping data
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
rank = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for d in rank:
    Rank.append(d.text)
name = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for d in name:
    Name.append(d.text)
art = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for d in art:
    Artist.append(d.text)
view = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for d in view:
    Views.append(d.text)
Date = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for d in Date:
    Upload_Date.append(d.text)
    
```

```python
# creating the dataframe from the scraped data
import pandas as pd
Videos=pd.DataFrame({})
Videos['Rank']=Rank[0:30]
Videos['Name']=Name[0:30]
Videos['Artist']=Artist[0:30]
Videos['Upload_Date']=Upload_Date[0:30]
Videos['Views']=Views[0:30]
Videos
```

# Question-2
Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details:
A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Opening the url
driver.get('https://www.bcci.tv/')
```

```python
#navigating to the international fixtures page from homepage through code
international_fixture=driver.find_element_by_xpath("//ul[@class='navigation__list showMoreEnabled']/li[1]/div[2]/div/ul/li[1]/a")
driver.get(international_fixture.get_attribute('href'))
```

```python
# creating empty lists for scraping data
Match_title=[] 
Series=[]
Place=[] 
Date=[] 
Time=[] 
```

```python
title = driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for d in title:
    Match_title.append(d.text)
series = driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]/span[2]')
for d in series:
    Series.append(d.text)
place = driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span')
for d in place:
    Place.append(d.text)
date = driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/span')
month = driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]/div/div/span[1]')
for d in range(len(date)):
    Date.append(date[d].text+month[d].text)
time = driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/div/span[2]')
for d in time:
    Time.append(d.text)
    
```

```python
# creating the dataframe from the scraped data
import pandas as pd
ICC=pd.DataFrame({})
ICC['Match_title']=Match_title
ICC['Series']=Series
ICC['Place']=Place 
ICC['Date']=Date
ICC['Time']=Time
ICC
```
